```python
from google.cloud import storage
import numpy as np
from io import BytesIO

def download_blob_as_numpy(bucket_name, blob_name):
    """Downloads a blob from GCS as a numpy array."""
    storage_client = storage.Client()
    bucket = storage_client.bucket(bucket_name)
    blob = bucket.blob(blob_name)
    
    byte_stream = BytesIO()
    blob.download_to_file(byte_stream)
    byte_stream.seek(0)
    
    return np.load(byte_stream)

# Example usage:
bucket_name = 'chess_data_all'
blob_names1 = ['bitboard_dataset_2-3M.npy', 'bitboard_dataset_5-6M.npy'] 
blob_names2 = ['labels_dataset_2-3M.npy', 'labels_dataset_5-6M.npy']# Add paths to all numpy arrays you want to fetch

arrays1 = [download_blob_as_numpy(bucket_name, blob_name) for blob_name in blob_names1]
arrays2 = [download_blob_as_numpy(bucket_name, blob_name) for blob_name in blob_names2]
```

```python
concatenated_array1 = np.concatenate(arrays1, axis=0)  # Adjust axis if needed
concatenated_array2 = np.concatenate(arrays2, axis=0)  # Adjust axis if needed
print(concatenated_array1.shape,concatenated_array1[20])
print(concatenated_array2.shape,concatenated_array2[20])
```

```python
def upload_numpy_to_gcs(bucket_name, numpy_array, destination_blob_name):
    """Uploads a numpy array to GCS as a byte stream."""
    storage_client = storage.Client()
    bucket = storage_client.bucket(bucket_name)
    blob = bucket.blob(destination_blob_name)
    
    byte_stream = BytesIO()
    np.save(byte_stream, numpy_array)
    byte_stream.seek(0)
    
    blob.upload_from_file(byte_stream, content_type='application/octet-stream')
    print(f"Numpy array uploaded to {destination_blob_name}.")

# Use the function
destination_blob_name1 = 'concatenated_array1-2,3-5,6.npy'
destination_blob_name2 = 'concatenated_array2-2,3-5,6.npy'
upload_numpy_to_gcs(bucket_name, concatenated_array1, destination_blob_name1)
upload_numpy_to_gcs(bucket_name, concatenated_array2, destination_blob_name2)
```
